```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_folder = r"C:\Users\geofr\Desktop\Research\Data\FE\PUND\Labview" 
path = data_folder + r"\HZO_V2\600PMA\E1_large.csv"
head = ['Pulse_width', 'rise_time', 'pulse_delay', 'voltage', 'p_u_charging', 'p_u_discharging', 'n_d_charging', 'n_d_discharging']
data = pd.read_csv(path, names=head)
```

```python
data['Pulse_width'].to_numpy()
```

```python
p_u_charging = data['p_u_charging'].to_numpy()
p_u_discharging = data['p_u_discharging'].to_numpy()
n_d_charging = data['n_d_charging'].to_numpy()
n_d_discharging = data['n_d_discharging'].to_numpy()
voltage = data['voltage'].to_numpy()
```

```python
plt.plot(voltage, p_u_charging,color='blue', label='p_u_charging')
plt.plot(voltage, n_d_charging, color='red', label='n_d_charging')
plt.plot(voltage, p_u_discharging, color='purple', label='p_u_discharging')
plt.plot(voltage, n_d_discharging, color='green', label='n_d_discharging')
plt.title("PUND - 2Pr vs V")
plt.xlabel("Voltage (V)")
plt.ylabel("2Pr")
plt.legend()
```

```python
p_diff = p_u_discharging - p_u_charging
n_diff = n_d_discharging - n_d_charging
```

```python
plt.plot(voltage, p_u_charging,color='blue')
plt.plot(-1*voltage, -1*n_d_charging, color='red')
plt.plot(voltage, p_u_discharging, color='purple')
plt.plot(-1*voltage, -1*n_d_discharging, color='green')
```

```python
plt.plot(voltage, n_d_charging, color='red')
plt.plot(voltage, n_d_discharging, color='green')
```
